```python
#code to adjust numbers in super module locator labelled output file, also to make images from just the output
#run this code to compare previous image and raw image to see if yaws line up to give the right smod labels
import os
import csv
import numpy as np
name = 'pmt_loc_label.txt'
out_name = 'fixed_pmt_loc_label.txt'
idir = '/home/dm3315/Documents/SK/PMT_learning/images/datasets/substacks_far/output/ring_avg_all/smodule/manual_intervention/'
img_info = []
img_nums = []
final_info = []

with open (f'{idir}{name}', 'r') as info :

    for row in csv.reader(info, delimiter=':') :
        img_num = int(row[0])
        pmt_x = float(row[1])
        pmt_y = float(row[2])
        label_1 = float(row[3])
        label_2 = float(row[4])
        label_3 = float(row[5])
        label_4 = float(row[6])
        img_nums.append(img_num)
        img_info.append([img_num, pmt_x, pmt_y, label_1, label_2, label_3, label_4])
img_info = np.array(img_info)
img_nums = np.unique(img_nums)
print(img_nums[0])
print(img_info[:, 0])
total_change = 0

pmt_label_file = open(f'{idir}{out_name}', 'w')
pmt_label_file.close()


for num in range(len(img_nums)) :
    pmt_label_file = open(f'{idir}{out_name}', 'a')
    current_img_info = img_info[np.equal(img_info[:, 0], int(img_nums[num]))]
    
    print(f'is this correct for {img_nums[num]} : {np.unique(current_img_info[:, 3])}')
    correction_needed= input("(y/n)\n-->")
    while (correction_needed.lower() == "n"):
        increase = input('Increase? (y/n)\n-->')
        if increase.lower() == 'y' :
            for i in current_img_info :
                #print(i)
                i[3] += 1
                total_change += 1
        else :
            for i in current_img_info :
                i[3] -= 1          
                total_change -= 1

        print(f'is this correct for {img_nums[num]} : {np.unique(current_img_info[:, 3])}')
        correction_needed= input("(y/n)\n-->")
    #final_info.append(current_img_info)
    #print(current_img_info)
    print('It Is Written')
    for j in current_img_info :
        #print(j)
        pmt_label_file.write(f'{int(j[0])}:{j[1]}:{j[2]}:{j[3]}:{j[4]}:{j[5]}:{j[6]} \n')
    pmt_label_file.close()
```
